$\text {5. Assuming no payments are made, how much would a \$1,000 be worth at 5% interest compounded monthly after 3 years?}$
Ans:
  Say- 
  * A = balance
  * P = Starting investment($1000)
  * r = Interest rate (5%)
  * t = Number of years (3)
  * n = Periods (12)
  \begin{equation} A = P × (1+ \frac {r}{n})^{nt} \end{equation}
    \begin{equation} A = 1000 × (1+ \frac {0.05}{12})^{12×3} = ? \end{equation}


```python
p, r , n, t = symbols('p r n t')
a = p * (1+(r/n))**(n*t)
a.evalf(subs={p:1000,r:5/100, n:12, t:3})
```

$\text{7. For the function }f(x) = 3x^2+1, \text{what is the slope at x=3?}$

```python
f, x = symbols('f x')
f =  3*(x**2)+1
dx_f=diff(f)
dx_f.subs(x,3)
```
