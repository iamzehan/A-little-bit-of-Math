# **`numpy.random.normal(loc, scale, size)`**
Is used to create normally distributed data when plotted looks like a bell curve.

<img src='https://cdn.scribbr.com/wp-content/uploads/2020/10/normal-distribution.png' width='400'>

* loc is the mean, default=0
* scale is the standard daviation; default=1
* size is number of instances we want to draw from the normally distributed data, it can be 1D,2D, 3D arrays. But if we put just one number here, the output will be a 1D array. If we put size=(3,3) then the output will be shown as a 3 × 3 np.ndarray, in which the data are drawn from the normally distributed data. 

In the example below we are drawing 100 data points, from a normally distributed data, that has a mean 10, standard daviation 5.
The random seed 439, draws an array of random numbers exclusive to the number 439. What this means is that if you provide the same seed, you will get the same output. And if you change the seed, you will get a different output.

````python
import numpy as np

def standardization(x):
    x_new= x-x.mean()/x.std()
    return x_new
np.random.seed(439) # 439 is just an input to generate pseudo-random numbers and these numbers specific to 439, if we input other numbers, those will be specific to those numbers only
x= np.random.normal(10, 5, 100) 
print(x.mean(), x.std())
"""parameters are loc(mean), scale(standard daviation), size(how many instances to draw from the distribution)"""
scaled_data= standardization(x)
print(len(scaled_data))
````

#1.2.3.2 Paris Apartments
___

We will take an example of modeling Paris apartment prices. 

*We want to find a way to predict the price of a Paris Apartment from several parameters Θ.

* We try to find a rule to weight the two features(square meters and numeber of rooms) to predict the price. However, The link between the features and  the price can be written as an equation:

\begin{equation}
 a × area + b × room = price
\end{equation}

The variables $a$ and $b$ are telling us the importance of each feature. A large positive value of $a$ means that the area has a big impact on the price, with large values corresponding to the large prices. A large negative value indicates an opposite relation, with large areas corresponding to small prices.

We start by looking at two apartments:

| Square Meters | Number of Rooms | Price |
|---------------|-----------------|-------|
|42             | 2               | 545000|
|39             | 1               | 495000|



````python
from sklearn.linear_model import LinearRegression

x= np.array([[42,2], [39,1]])
y= np.array([	545000,495000])
````

````python
lr= LinearRegression()
lr.fit(x,y)
````

````python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False) #copy_X=True The input X, when you try to fit/etc the model will, if the flag is set to true (default), be copied for use within the function. This means that the original X you passed as a parameter will be the same after the fit/etc has been performed.
````

````python
import matplotlib.pyplot as plt
plt.scatter(x[:,1],y)
plt.plot(x[:,1], lr.predict(x))
````

 In linear regression, coefficients are the values that multiply the predictor values. Suppose you have the following regression equation: 
 \begin{equation}
 y = 3X + 5. 
 \end{equation}
 In this equation, $+3$ is the coefficient, $X$ is the predictor, and $+5$ is the constant.

````python
print(lr.coef_)
print(lr.intercept_)
````

\begin{equation}

 (15000 × area ) + (5000 × room) - 95000 = price

It looks like the rule is perfect! Not so fast! Does it mean that our model is good? Let's try with a third apartment that we didn't use to create our model. Here's the data:

<h2 align="center"> <strong> Using limits to calculate Derivatives. </strong> </h2>

___

Use a limit to forever decrease a step $s$ and see what slope it approaches:

\begin{equation}
\lim_{s\to 0} \frac {(x + s)^2 - x^{2}}{(x + s) - x}
\end{equation}
\begin{equation}
\lim_{s\to 0} \frac {(2 + s)^2 - 2^{2}}{(2 + s) - 2}
\end{equation}


Here's how to implement it in `SymPy`:

````python
from sympy import *
x, s  = symbols('x s')

f = x**2

slope_f = ( f.subs(x , x+s ) -f )/ ((x+s) - x)

# substitute 2 for x
slope_2 = slope_f.subs(x,2)

#infinitely approach step size s to 0
result =limit(slope_2,s,0)
````

### **Integrals**

Simply, the opposite of derivative is $integral$, which finds the area under the curve in a given range.

<p align="center">
<img src="https://matplotlib.org/3.1.0/_images/sphx_glr_integral_001.png">
</p>

````python
def approximate_integral(a, b, n, f):
  delta_x = (b-a)/n
  total_sum = 0
  for i in range(1,n+1):
    midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
    total_sum += f(midpoint)
  return total_sum * delta_x
def my_function(x):
  return x**2 + 1

area = approximate_integral(a=0, b= 1, n = 1000, f = my_function)
# for 1000 points, ranging between 0 to 1, find the area under the curve for the function f(x)=x**2+1
````

````python
print(area)
````

no matter the number of $n$ the result would be the same.

### **Calculating Integrals with Limits**

````python
x, i, n = symbols('x i n')

f = x**2 + 1
lower, upper =0,1

#Calculate width and each rectangle height at index "i"
delta_x= ((upper-lower)/n) #width
x_i = (lower + delta_x *i) 
fx_i =f.subs(x,x_i) # height

# Iterate all "n" rectangles and sum their areas
n_rectangles = Sum(delta_x* fx_i, (i, 1, n)).doit()

area = limit (n_rectangles, n, oo)
display(area)
````
