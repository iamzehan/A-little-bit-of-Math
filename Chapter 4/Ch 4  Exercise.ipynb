### **Exercise 1**
---
Vector $\vec{v}$ has a value of $[1,2]$ but then a transformation happens. $\hat{i}$ lands at $[2,0]$ and $\hat{j}$ lands at $[0,1.5]$. Where does $\vec{v}$ land?

```python
v=array([1,2])
i_hat = array([2,0])
j_hat = array([0, 1.5])
#fix this line 
basis = array([i_hat, j_hat])

# transform vector v into w 
w = basis.dot(v)
print(w)
```

#### **Exercise 3**
---
A transformation $\hat{i}$ lands at $[1,0]$ and $\hat{j}$ lands at $[2,2]$. What is the determinant of this transformation?

```python
i_hat = array([1,0])
j_hat = array([2,2])
basis = array([i_hat, j_hat]).transpose()
determinant = det(basis)
print(determinant)
```

#### **Exercise 5**

Solve the system equations for x,y and z:
\begin{equation}
3x + 1y + 0z = 54\\
2x + 4y + 1z = 12\\
3x + 1y + 8z = 6
\end{equation}

```python
A = array([
           [3,1,0],
           [2,4,1],
           [3,1,8]
])
B = array([
           54,
           12,
           6
]) 
X = inv(A).dot(B)
print(X)
```
