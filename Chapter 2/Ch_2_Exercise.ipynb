#**Exercise 1:**

There is a 30% chance of rain today, and a 40% chance your ubrella order will arive on time. You are eager to walk in the rain today and cannot do so without either!

What is the probability it will rain AND your umbrella will arrive?

\begin{equation}
P(Rain\space AND\space Umbrella) = {P(Umbrella)\times P(Rain)}
= 0.4 × 0.3 = 0.12
\end{equation}

# **Exercise 2:**
There is a 30% chance of rain today, and a 40% chance your ubrella order will arive on time. You will be able to run errands only if it doesn't rain or your umbrella arrives.
What is probability it will not rain OR your umbrella arrives?

\begin{align}
P(NR \space \text{OR} \space U) = P(NR) + P(U) - P(NR \space AND \space U) \\
P(NR \space \text{OR} \space U) = (1.0 - 0.3) + 0.4 - 0.12 = 0.7 + 0.4 - 0.12 = 0.98
\end{align}

# Exercise 5:
You flipped a coin 19 times and got heads 15 times and tails 4 times.
Do you think this coin has any good probability of being fair? Why and why not?

```python
from scipy.stats import beta

a = 15
b = 4

dist= 1.0 - beta.cdf(0.5,a,b)
print(dist)
```
